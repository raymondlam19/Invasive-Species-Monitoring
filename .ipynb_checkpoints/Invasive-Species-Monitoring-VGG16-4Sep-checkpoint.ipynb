# Load dataset

```python
data = np.load('./input/256x256.npz')
data.files
```

```python
x_train = data['x_train']
y_train = data['y_train']
x_test = data['x_test']
```

# Split training and valdiation set

```python
from sklearn.model_selection import train_test_split

x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, test_size = 0.1, random_state=0)

print('len(x_train):', len(x_train))
print('len(x_val):', len(x_val))
print('len(y_train):', len(y_train))
print('len(y_val):', len(y_val))
```

# Keras pre-trained VGG16

```python
from keras.models import Sequential, Model, load_model
from keras import applications
from keras import optimizers
from keras.layers import Dropout, Flatten, Dense, GlobalAveragePooling2D
```

```python
# Define model architecture

def Create_model():
    
    image_shape = (256,256,3)
    base_model = applications.VGG16(weights='imagenet', include_top=False, input_shape=image_shape)
    base_model.trainable = False
    
    model = Sequential()
    model.add(base_model)
    
    # model.add(Flatten())
    model.add(GlobalAveragePooling2D())
    model.add(Dropout(0.25))
    
    # model.add(Dense(256, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))

    return model
```

```python
# Create model

model = Create_model()
model.summary()
```

```python
# Compile the model

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
# Data augmentation (ImageDataGenerator)

from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
                    rotation_range=30, 
                    width_shift_range=0.1,
                    height_shift_range=0.1, 
                    horizontal_flip=True)

train_datagen.fit(x_train)
```

```python
# Define callbacks (EarlyStopping, ReduceLROnPlateau, ModelCheckpoint)
# ModelCheckpoint: Save the (best) model weights after every epoch

from keras.callbacks import EarlyStopping, ReduceLROnPlateau, ModelCheckpoint

callbacks = [EarlyStopping(monitor='val_loss', patience=2, min_delta=1e-2),
             ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=1, cooldown=1, min_lr=1e-5),
             ModelCheckpoint(filepath='./model/VGG16-transferlearning.model', monitor='val_acc',
                             save_best_only=True, save_weights_only=False)]
```

```python
# Train the model

batch_size = 32
epochs = 10

history = model.fit_generator(train_datagen.flow(x_train, y_train, batch_size=batch_size),
                              steps_per_epoch=x_train.shape[0] // batch_size, 
                              epochs=epochs, 
                              validation_data=(x_val,y_val),
                              callbacks=callbacks)
```

```python
print(history.history.keys())
```

# Model Evaluation (df_train)

## 1) Plot loss and accuracy (train, val)

```python
plt.plot(history.history['acc'])
plt.plot(history.history['val_acc'])
plt.xlabel('Epoch')
plt.legend(['train', 'dev'], loc='upper left')
plt.title('Model accuracy')
plt.ylabel('accuracy')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'dev'], loc='upper left')
plt.show()
```

## 2) Loss and accuracy (val)

```python
score = model.evaluate(x_val, y_val)
print("loss:", score[0])
print("metric (accuracy): ", score[1])
```

## 3) ROC AUC (val)

```python
y_val_pred = model.predict_classes(x_val)
```

```python
print('y_val_pred.shape:', y_val_pred.shape)
y_val_pred
```

```python
from sklearn.metrics import roc_auc_score

print('ROC AUC (val): {0}'.format(roc_auc_score(y_val, y_val_pred)))
```

## 3) Plot confusion matrix (val)

```python
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(y_true, y_pred, class_names, vmax=None, normed=True, title='Confusion matrix'):
    matrix = confusion_matrix(y_true,y_pred)
    if normed:
        matrix = matrix.astype('float') / matrix.sum(axis=1)[:, np.newaxis]
    fig = sns.heatmap(matrix, vmax=vmax, annot=True, square=True,
                cmap=plt.cm.Blues_r, cbar=False, linecolor='black', linewidths=1,
                xticklabels=class_names, yticklabels=class_names)
    
    plt.xlabel("Predicted labels")
    plt.ylabel("True labels")
    plt.title(title)
    plt.show(fig)
```

```python
plot_confusion_matrix(y_val, y_val_pred,
                      class_names=['Positive', 'Negative'],
                      title='Confusion matrix (val data)')
```

## 4) Plot errors

```python
errors = (np.squeeze(y_val_pred) - y_val != 0) # array of bools with true when there is an error or false when the image is cor

y_val_pred_er = y_val_pred[errors]
y_val_er = y_val[errors]
x_val_er = x_val[errors]

def display_errors(errors_index, x_val_er, y_val_pred_er, y_val_er):
    """ This function shows 6 images with their predicted and real labels"""
    nrows = 2
    ncols = 3
    
    fig, ax = plt.subplots(nrows, ncols, sharex=True, sharey=True, figsize=(12,10))
    for row in range(nrows):
        for col in range(ncols):
            i = row * ncols + col
            
            ax[row,col].imshow(x_val_er[i], )
            ax[row,col].set_title("Predicted label :{}\nTrue label :{}".format(y_val_pred_er[i],y_val_er[i]))

# Show the 6 errors
display_errors(errors_index, x_val_er, y_val_pred_er, y_val_er)
```
